The data comes in the form of a csv file from a police check point.

```python
import pandas as pd
import numpy as np
```

```python
police_df = pd.read_csv("3. Police Data.csv")
police_df.head(100)
```

```python
### Country_name column seems to contain null values only 

police_df.isnull().sum()
```

Country Name column only contains null values, column must be removed


```python
police_df.drop(columns = 'country_name', inplace = True)
police_df.columns
```

Questions for this dataset, does a persons gender influence their driving habits( who speeds more), and does gender affect who gets searched during a stop? 

```python
#For speeding, were Men or Women stopped more often?
police_df.head()
```

```python
police_df[police_df.violation == 'Speeding'].driver_gender.value_counts()
```

```python
##

police_df.groupby("driver_gender").search_conducted.sum()
```

```python
police_df.search_conducted.value_counts()
```

Men speed more than Women, Also they are searched more at stops.

```python
###Finding the mean stop_duration
police_df.head()
```

```python
police_df['stop_duration'].value_counts()
```

```python
#Since we have a range of stop duration not exact number of minutes I'll map them into the middle point

police_df['stop_duration'] = police_df['stop_duration'].map({'0-15 Min' : 7.5, '16-30 Min' : 23.5, '30+ Min' : 45})
```

```python
police_df.head()
```

```python
police_df['stop_duration'].mean()
```

On average a stop lasts about 12 minutes.
