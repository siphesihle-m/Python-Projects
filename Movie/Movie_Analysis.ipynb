```python
###LIbraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
plt.style.use("ggplot")
from matplotlib.pyplot import figure

#Adjusting the configuration of the plots we will create
%matplotlib inline
matplotlib.rcParams["figure.figsize"]=(12,8)
```

```python
##Reading in the data
movie_df = pd.read_csv("movies.csv")
```

```python
#checking data
movie_df.head()
```

```python
movie_df.shape
```

DATA CLEANING

```python
### Finding the missing value percentage of each coolumn

for col in movie_df.columns:
    pct_missing = np.mean(movie_df[col].isnull())
    print(f"{col} - {pct_missing}%")
```

```python
###Small missing values highest is budget with 28%
#Dropping missing values
movie_df =movie_df.dropna()



```

```python
movie_df.shape
```

```python
##checking the data types of each column
movie_df.dtypes
```

```python
###changing gross and budget data type

movie_df["gross"] = movie_df["gross"].astype("int64")
movie_df["budget"] = movie_df["budget"].astype("int64")
```

```python
movie_df.head()
```

```python
### create correct year released column
dates = np.array(movie_df["released"].astype(str))
correct_date = np.array([])
for date in dates:
    correct_date = np.append(correct_date,date.split("(")[0][-2:-6:-1])
movie_df['Corrected year'] = correct_date
```

```python
movie_df["Corrected year"] =movie_df["Corrected year"].astype(str).str[::-1]
```

```python
movie_df["Corrected year"] = movie_df["Corrected year"].astype(int)
movie_df["Corrected year"]
```

```python
### sort data by descending gross
movie_df.sort_values(by=["gross"], inplace= False, ascending=False)
```

```python
movie_df.drop_duplicates()
### data doesn't have any duplicatess
```

Analysis

```python
#### My theory 
#Budget high correlation to revenue
#Company high correlation to revenue
```

```python
# Testing the hypothesis

#Scatter plot with budget vs gross

plt.scatter(x=movie_df["budget"], y=movie_df["gross"])
plt.title("Budget vs Gross Earnings")
plt.xlabel("Budget for Film")
plt.ylabel("Gross Earnings")
plt.show()
```

```python
###checking for correlation (budget vs gross) using regression plot

sns.regplot(x = "budget", y="gross", data=movie_df, scatter_kws={"color":"red"},line_kws={"color":"blue"})
```

```python
### The regression line shows us that the correlation is positive 

movie_df.corr()
```

```python
###High correlation between budget and gross
### My hypothesis was right
```

```python
correlation_matrix = movie_df.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation Matrix for Numeric Features")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
plt.show()
```

```python
#### Looking at the correlation of company a string variable

movie_df.head()
```

```python
###change object columns to categorical variables

temp_df = movie_df
```

```python
for col_name in temp_df.columns:
    if temp_df[col_name].dtypes== 'object':
        temp_df[col_name] = temp_df[col_name].astype("category")
        temp_df[col_name] = temp_df[col_name].cat.codes
```

```python
temp_df.head()
```

```python
correlation_matrix = temp_df.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation Matrix for Numeric Features")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
plt.show()
```

```python
corr_pairs =correlation_matrix.unstack()
print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs)>0.5]
```
